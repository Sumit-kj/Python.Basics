<h1>Python - 105 - Iterations</h1>

<p>A loop statement allows us to execute a statement or group of statements multiple times. The following diagram illustrates a loop statement.</p>

<h3>while loop</h3>
<p>A while loop statement in Python programming language repeatedly executes a target statement as long as a given condition is true.</p>
<h4>Syntax</h4>
<p>while expression:</p>
<p>&nbsp;&nbsp;&nbsp;&nbsp;statement(s)</p>

```python
count = 0
while (count < 9):
    print ('The count is:', count)
    count = count + 1
print ("Good bye!")
```

<h4>else with while loop</h4>
<p>the else statement is executed when the condition becomes false</p>

```python
count = 0
while count < 5:
    print (count, " is less than 5")
    count = count + 1
else:
    print (count, " is not less than 5")
```

<h3>for loop</h3>
<p>The for statement in Python has the ability to iterate over the items of any sequence, such as a list or a string.</p>
<h4>Syntax</h4>
<p>for iterating_var in sequence:</p>
<p>&nbsp;&nbsp;&nbsp;&nbsp;statement(s)</p>

```python
for letter in 'Python':
    print ('Current Letter :', letter)
print()
fruits = ['banana', 'apple', 'mango']
for fruit in fruits:
    print ('Current fruit :', fruit)
print ("Good bye!")
```

<h4>range() function</h4>
<p>range(n) generates an iterator to progress integers starting with 0 upto n-1.</p>

```python
for var in list(range(5)):
    print(var)
```

```python
fruits = ['banana', 'apple', 'mango']
for index in range(len(fruits)):
    print ('Current fruit :', fruits[index])
print ("Good bye!")
```

<h4>else with for loop</h4>
<p>the else block is executed only if for loops terminates normally (and not by encountering break statement).</p>

```python
numbers=[11,33,54,39,55,75,37,21,23,41,13]
for num in numbers:
    if num%2==0:
        print ('the list contains an even number')
        break
else:
    print ('the list does not contain even number')
```

<h3>Nested Loops</h3>
<p>Python programming language allows the use of one loop inside another loop.</p>
<h4>Syntax for nested for</h4>
<p>for iterating_var in sequence:</p>
<p>&nbsp;&nbsp;&nbsp;&nbsp;for iterating_var in sequence:</p>
<p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;statements(s)</p>
<p>&nbsp;&nbsp;&nbsp;&nbsp;statements(s)</p>
<h4>Syntax for nested while</h4>
<p>while expression:</p>
<p>&nbsp;&nbsp;&nbsp;&nbsp;while expression:</p>
<p>&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;&nbsp;statements(s)</p>
<p>&nbsp;&nbsp;&nbsp;&nbsp;statements(s)</p>

```python
for i in range(1,11):
    for j in range(1,11):
        k=i*j
        print (k, end=' ')
    print()
```

<h3>Loop control statements</h3>
<p>break statement - Terminates the loop statement and transfers execution to the statement immediately following the loop.</p>
<p>continue statement - Causes the loop to skip the remainder of its body and immediately retest its condition prior to reiterating.</p>
<p>pass statement - The pass statement in Python is used when a statement is required syntactically but you do not want any command or code to execute.</p>

<h3>break statement</h3>
<p>The break statement is used for premature termination of the current loop. After abandoning the loop, execution at the next statement is resumed.</p>
<p>If you are using nested loops, the break statement stops the execution of the innermost loop and starts executing the next line of the code after the block.</p>

```python
for letter in 'Python':
    if letter == 'h':
        break
    print ('Current Letter :', letter)
```

```python
var = 10
while var > 0:
    print ('Current variable value :', var)
    var = var -1
    if var == 5:
        break
print ("Good bye!")
```

<h3>continue statement</h3>
<p>The continue statement in Python returns the control to the beginning of the current loop. When encountered, the loop starts next iteration without executing the remaining statements in the current iteration.</p>
<p>If you are using nested loops, the break statement stops the execution of the innermost loop and starts executing the next line of the code after the block.</p>

```python
for letter in 'Python':
    if letter == 'h':
        continue
    print ('Current Letter :', letter)
```

```python
var = 10
while var > 0:
    var = var -1
    if var == 5:
        continue
    print ('Current variable value :', var)
print ("Good bye!")
```

<h4>pass statement</h4>
<p>It is used when a statement is required syntactically but you do not want any command or code to execute. It is like a checkpoint</p>
<p>The pass statement is a null operation; nothing happens when it executes. The pass statement is also useful in places where your code will eventually go, but has not been written yet i.e. in stubs).</p>

```python
for letter in 'Python':
    if letter == 'h':
        pass
        print ('This is pass block')
    print ('Current Letter :', letter)
print ("Good bye!")
```
